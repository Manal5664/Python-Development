# List

```python
# Creating a list
fruits = ["apple", "banana", "mango"]

# Accessing items
print(fruits[2])        # Output: apple

# Adding a new item
fruits.append("grape")
# print(fruits)

# # Removing an item
fruits.remove("banana")
print(fruits)

# # Displaying the updated list
# print(fruits)           # Output: ['apple', 'mango', 'grape']

# fruits.append("grape")

# print(fruits)           # Output: ['apple', 'mango', 'grape']
```

```python
# Creating a tuple
colors = ("red", "green", "blue" ,'red')
# Accessing items
print(colors[3])         # Output: green

# Trying to change a tuple (will cause error)
# colors[0] = "yellow"   
# But you can loop through a tuple
# for c in colors:
#     print(c)



# Ordered

# Cannot be changed

# Allows duplicates
```

```python
# Creating a set
numbers = {1, 2, 3, 3, 2, 4, 4}
# Displaying the set
print(numbers)           # Output: {1, 2, 3, 4}  (duplicates removed)

# Adding new item
numbers.add(5)

# Removing item
numbers.remove(2)

# Displaying updated set
print(numbers)           # Output: {1, 3, 4, 5}


# Unordered

# No duplicates

# Can add/remove items
```

```python
# Creating a dictionary
student = {
    "name": "Asif",
    "age": 21,
    "marks": 88
}

print(student)

# Accessing a value using key
print(student["age"])        # Output: Ali

# Adding a new key–value pair
student["grade"] = "A"

# Updating a value
student["marks"] = 92

# Removing a key–value pair
student.pop("age")

# Displaying the updated dictionary
print(student)


# Data stored in key–value pairs

# Keys must be unique

# Values can be changed (mutable)
```
