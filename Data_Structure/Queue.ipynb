```python
queue = []

queue.append(10)
queue.append(20)
queue.append(30)

queue.pop(0)
print(queue)
```

```python
from collections import deque

queue = deque()

queue.append(10)     # enqueue
queue.append(20)
queue.append(30)

queue.popleft()      # dequeue
print(queue)
```

```python
from queue import Queue

q = Queue(maxsize=3)

q.put(10)
q.put(20)
q.put(30)

print(q.get())
```
