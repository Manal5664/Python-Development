```python
data = (10, 20, 30)
print(data)
```

```python
# Access
print(data[1])
```

```python
# Tuple Unpacking
a, b, c = data
print(a, b, c)
```

```python
t = (5,)
print(type(t))
```

```python
# Tuple element access
data = (100, 200, 300)
print(data[0])
print(data[-1])
```

```python
# Tuple slicing
print(data[0:2])
```

```python
# Common Methods of Tuple
nums = (1, 2, 3, 2, 2)

print(nums.count(2))
print(nums.index(3))
```

```python
# Loop in Tuple
nums = (1, 2, 3, 2, 2)
for item in nums:
    print(item)
```
