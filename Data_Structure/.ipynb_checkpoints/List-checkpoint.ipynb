```python
numbers = [10, 20, 30, 40]
print(numbers)
```

```python
# Access Element
print(numbers[0])
print(numbers[-1])
```

```python
# Modify list 
numbers[1] = 25
print(numbers)
```

```python
# Common list method 
numbers.append(50)
print(numbers)
```

```python
numbers.insert(1, 15) #insert(index, value)
print(numbers)
```

```python
numbers.remove(30)
print(numbers)
```

```python
numbers.pop()
print(numbers)
```

```python
numbers.sort()
print(numbers)
```

```python
# loop with list 
for num in numbers:
    print(num)
```
