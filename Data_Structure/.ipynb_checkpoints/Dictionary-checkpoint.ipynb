```python
student = {
    "name": "Manal",
    "age": 21,
    "course": "Python"
}
```

```python
# Access Value
print(student["name"])
print(student.get("age"))
```

```python
# Modify dictionary
student["age"] = 2
student["city"] = "Karachi"
```

```python
# Loop Dictionary
for key, value in student.items():
    print(key, value)
```

```python
# Delete Dictionary

del student["course"]
student.pop("age")
```

```python
# Loop ion Dict
for key in student:
    print(key, student[key])
```

```python
# Common methods in Dict
print(student.keys())
print(student.values())
print(student.items())
```
