```python
# Writing to a File

f = open("test.txt", "w")
f.write("Hello Python")
f.close()
```

```python
# Appending to a File
f = open("test.txt", "a")
f.write("\nLearning file handling")
f.close()
```

```python
# Reading from a File
f = open("test.txt", "r")
data = f.read()
print(data)
f.close()
```

```python
# Read Line by Line
f = open("test.txt", "r")
print(f.readline())
f.close()

```

```python
# Read All Lines as List
f = open("test.txt", "r")
lines = f.readlines()
print(lines)
f.close()
```

```python
# Closing a file
f.close()
```

```python
# Using with Statement (Best Practice)
with open("test.txt", "r") as f:
    data = f.read()
    print(data)
```

```python
# Checking if File Exists
import os

if os.path.exists("test.txt"):
    print("File exists")
else:
    print("File not found")
```

```python
# Deleting a File
import os
os.remove("test.txt")
```
