```python
hr=float(input('Enter Hours'))
Rate=float(input('Enter Rate'))
if hr>=40:
    reg_pay= 40* Rate
    overtime=(hr-40)*Rate*1.5
    pay=reg_pay+overtime
else:
    pay=print(float(hr*Rate))

print(pay)
```
