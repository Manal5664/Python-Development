```python
# Basic function examples

def greet():
    print("Hello, welcome to Python learning!")

greet()
```

```python
def add_numbers(a, b):
    return a + b
result = add_numbers(5, 3)
print("Sum:", result)


```

```python
def square(num):
    return num * num

print("Square of 4:", square(4))
```

```python
def greet_user(name):
    print("Hello", name)

greet_user("Manal")
```

```python
def multiply(a, b):
    return a * b

result = multiply(4, 5)
print("Result:", result)
```

```python
def welcome(name="User"):
    print("Welcome", name)

welcome()
welcome("Manal")
```

```python
def calculate(a, b):
    return a + b, a - b, a * b

add, sub, mul = calculate(10, 5)
print(add, sub, mul)
```

```python
def outer_function():
    def inner_function():
        print("Inner function called")
    inner_function()

outer_function()
```

```python
def add_numbers():
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    print("Sum:", a + b)

add_numbers()
```

```python
def check_age(age):
    if age >= 18:
        return "Adult"
    else:
        return "Minor"

print(check_age(15))
```
