```python
num1=int(input("Enter number 1:"))
num2=int(input("Enter number 2:"))
opr=input('Enter the operator:')

if opr=='+':
    print(num1+num2)
elif opr=='-':
    print(num1-num2)
elif opr=='*':
    print(num1*num2)
elif opr=='/':
    print(num1/num2)
else:
    print('Invalid Operation')
    
```
