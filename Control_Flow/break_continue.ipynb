```python
# break example
for i in range(1, 10):
    if i == 5:
        break
    print(i)

print("Loop stopped")
```

```python

# continue example
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

```python
# Example 1: break – stop loop when number found
numbers = [2, 4, 6, 8, 10, 7, 12]

for num in numbers:
    if num == 7:
        print("Number found, stopping loop")
        break
    print(num)
```

```python
# Example 2: continue – skip even numbers
for i in range(1, 11):
    if i % 2 == 0:
        continue
    print(i)
```

```python
# Example 3: break in while loop
count = 1

while count <= 10:
    if count == 6:
        break
    print(count)
    count += 1
```

```python
# Example 4: continue in while loop
count = 0

while count <= 5:
    count += 1
    if count == 3:
        continue
    print(count)
```
