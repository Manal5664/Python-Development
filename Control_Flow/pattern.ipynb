```python
# Pattern printing using nested loops

# Star pattern
rows = 5

for i in range(1, rows + 1):
    for j in range(i):
        print("*", end=" ")
    print()

```

```python

# Number pattern
for i in range(1, rows + 1):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

```python
# reverse patter
rows = 5

for i in range(rows, 0, -1):
    for j in range(i):
        print("*", end=" ")
    print()
```

```python
# pyramid pattern
rows = 5

for i in range(1, rows + 1):
    print(" " * (rows - i), end="")
    print("* " * i)
```
