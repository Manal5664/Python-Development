## Example 1: Even or Odd

```python
num = float(input("Enter first number: "))


if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

## Example 2: Eligible for Voting

```python
age = 20

if age >= 18:
    print("Eligible to vote")
else:
    print("Not eligible to vote")
```

## Nested if- else

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operation = input("Enter operation (+, -, *, /): ")
if operation == "+":
    result = num1 + num2

elif operation == "-":
    result = num1 - num2


elif operation == "*":
    result = num1 * num2

elif operation == "/":
    if num2 != 0:
        result = num1 / num2
    else:
        result = "Error: Division by zero is not allowed."

else:
    result = "Error: Invalid operation."

print(f"Result: {result}")
```

# Example 1: Largest of Two Numbers (using nested if)

```python
a = 10
b = 20

if a != b:
    if a > b:
        print("a is greater")
    else:
        print("b is greater")
else:
    print("Both are equal")
```

## Example 2: Login System

```python
username = "admin"
password = "124"

if username == "admin":
    if password == "1234":
        print("Login successful")
    else:
        print("Wrong password")
else:
    print("Wrong username")
```
